```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

text = "Data engineering is the process of discovering, designing and building the data infrastructure to help data owners and data users use and analyze raw data from multiple sources and formats"
wordcloud= WordCloud( width= 800,height=400, background_color= 'black',prefer_horizontal=1).generate(text)
plt.figure(figsize=(10,5))
plt.imshow(wordcloud,interpolation ='bilinear')
plt.axis("off")
plt.show()
```
