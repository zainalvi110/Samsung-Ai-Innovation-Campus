## Activity 2 — Split and Merge Text Columns
**Goal:** Learn to split, merge, and create new textual features.

```python
# TODO: Extract email domain from 'Email' column into a new column 'Email_Domain'

# Extract email domain from 'Email' column
df['Email_Domain'] = df['Email'].str.split('@', expand=True)[1]

#[0] → username part (before the @)
#[1] → domain part (after the @)
#[1] selects the domain part only.

#expand=True tells pandas to return a DataFrame (so you can select [1] cleanly).

# Quick check
df[['Email', 'Email_Domain']].head()
```

```python
# TODO: From 'Address', create two columns: 'City' and 'Province'
# Example: "Lahore, Punjab" -> City='Lahore', Province='Punjab'


# Split the 'Address' column into 'City' and 'Province'
df[['City', 'Province']] = df['Address'].str.split(',', n=1, expand=True)


#.str.split(',', n=1, expand=True) splits each address only once at the first comma.
#expand=True returns a DataFrame (so we can assign directly to two new columns).
#.str.strip() removes any extra spaces (important for neat results).

# Clean up whitespace (optional but good practice)
df['City'] = df['City'].str.strip()
df['Province'] = df['Province'].str.strip()

# Quick check
df[['Address', 'City', 'Province']].head()
```

### Step: Drop Redundant or Sensitive Columns
